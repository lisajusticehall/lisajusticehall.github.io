## 1. Healthcare Data In Context
<p>Valuable information can be extracted from healthcare data when groups of people share similar conditons. Medical conditons such as obsessive compulsive disorder(OCD) could have common variables, and using anonymized data we can determine what risk factors could increase the chancce for an individual to develop OCD. </p>
<p>In this example, we are going to use SQL in order to extract information from a dataset of individuals who have the medical characteristics of OCD. Their information has been anonmyized. We hope to answer the following questions: </p>
<ul>
<li>Does gender increase the risk for developing or having OCD?</li>
<li>Do certain ethnicities have an elevated the risk for developing or having OCD?</li>
<li>What are the most common co-occuring mental health diagnosis?</li>
<li>What is the most common obsession type and count?</li>
</ul>
<p>The first line of code, should allow us to have a preliminary view of what type of information is in the current database. We have information such as gender, ethnicity, currecnt medications, and more related data that has been collected. </p>

```python
%%sql

SELECT * 
FROM ocd.ocd_patient_dataset 
LIMIT 10
```

## 2. Ordering and Grouping Data
<p>Using COUNT and AVG functions, we can use the data we have to count the number of female and male patients who have OCD. We can expand further to calculate the average obsession score for each gender.</p>

```python
%%sql

SELECT Gender,
COUNT(`Patient ID`) as Patient_Count,
AVG(`Y-BOCS Score (Obsessions)`) as Avg_Obs_Score
FROM ocd.ocd_patient_dataset
GROUP BY Gender
ORDER BY Patient_Count;
```

<p>We can further investigate the data using the Count and Average functions to seperate the patients by ethnicity. </p>

```python
%%sql

SELECT Ethnicity,
COUNT(`Patient ID`) as Patient_Count,
AVG(`Y-BOCS Score (Obsessions)`) as Avg_Obs_Score
FROM ocd.ocd_patient_dataset
GROUP BY Ethnicity
ORDER BY Patient_Count;
```

<p>Digging deeper still into the data, we can determine which previous diagnosis are connected to Obessive Compulsive Disorder. We can also average the obsession score for each of those diagnosis. 

```python
%%sql

SELECT `Previous Diagnoses`,
COUNT(`Patient ID`) as Patient_Count,
AVG(`Y-BOCS Score (Obsessions)`) as Avg_Obs_Score
FROM ocd.ocd_patient_dataset
GROUP BY `Previous Diagnoses`
ORDER BY Patient_Count;
```
